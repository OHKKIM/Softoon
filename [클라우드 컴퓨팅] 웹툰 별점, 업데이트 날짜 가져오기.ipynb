```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
html=urlopen("https://comic.naver.com/webtoon/list.nhn?titleId=720121&weekday=mon")
bsObj=BeautifulSoup(html, "html.parser") #웹 페이지의 소스 코드 전부 가져옴
```

```python
bsObj
```

```python
#별점 정보
allStar=bsObj.findAll("span",{"class":"star"})
```
