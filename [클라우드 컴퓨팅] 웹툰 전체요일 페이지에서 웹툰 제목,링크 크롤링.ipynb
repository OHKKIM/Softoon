```python
import requests
from bs4 import BeautifulSoup

UP_TOTAL_LIST='https://comic.naver.com/webtoon/weekday.nhn'
response=requests.get(UP_TOTAL_LIST)
```

```python
response
```

```python
#웹툰 메인 페이지의 소스
response.text
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup
```

```python
#<a href~ 형식은 모두 가져옴
class_title_a_list=soup.select('a.title')
```

```python
class_title_a_list
```

```python
title_list = []

for a in class_title_a_list:
    a_href=a.get('href') #속성을 인자로 해서 해당 속성의 속성값을 가져온다??
    a_text=a.get_text() #태그 안의 내용을 가져옴
    result=f'{a_text}(https://comic.naver.com/{a_href})'
    title_list.append(result)
```

```python
print('\n'.join(title_list))
```
